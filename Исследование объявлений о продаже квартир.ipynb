##### Столбец  `floors_total`

Приступим к обработке столбца `floors_total`. Полагая, что данный столбец тоже заполнялся пользователем, можно сделать вывод, что пропуски в данных - результат человеческого фактора. Доля пропущенных значений в датафрейме крайне незначительна, а так как мы не можем определить по имеющимся данным, является ли этаж продаваемой квартиры последним, а для приведения к целочисленным значениям столбца, необходимо избавиться от прокусков данных, удалим строки с пропусками в столбце `floors_total` из датафрейма.

```python
print('Уникальные значения по столбцу \'floors_total\':\n',data['floors_total'].unique())
print()
print('Доля пропущенных значений по столбцу  \'floors_total\':',data['floors_total'].isna().sum())
```

```python
data=data.dropna(subset=['floors_total']).reset_index(drop=True)
data['floors_total']=data['floors_total'].astype('int')
print('Уникальные значения по столбцу \'floors_total\':\n',data['floors_total'].unique())
print()
print('Доля пропущенных значений по столбцу  \'floors_total\':',data['floors_total'].isna().sum())
```

##### Столбец  `is_apartment`

Столбцом с самым большим количеством пропущенных значения является столбец `is_apartment`. Предположим, что пропуски в значениях столбца означают буллево значение `False`, поскольку данный столбец, скорей всего заполнялся вручную пользователем, то отсутствующее значение может означать, что жилье аппартаментами не является. Произведем замену. 

```python
data['is_apartment'].value_counts()
```

```python
data['is_apartment'].fillna(False, inplace=True)
data['is_apartment'].value_counts()
```

```python
dict1 = {True: 1, False: 0}
data['is_apartment']=data['is_apartment'].map(dict1)
data['is_apartment'].value_counts()
```
